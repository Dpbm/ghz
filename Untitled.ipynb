```python
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector
from qiskit.transpiler.preset_passmanagers import generate_preset_pass_manager
from qiskit_aer import AerSimulator
from qiskit_ibm_runtime import QiskitRuntimeService
```

```python
service = QiskitRuntimeService()
backend = service.backend("ibm_osaka")
fake_backend = AerSimulator.from_backend(backend)
transpiler = generate_preset_pass_manager(optimization_level=3, backend=fake_backend)
```

```python
depth = lambda x: len(x.qubits) == 2
```

```python
ghz = QuantumCircuit(11)
ghz.h(5)
ghz.cx(5,4)
ghz.cx(5,6)

ghz.cx(4,3)
ghz.cx(3,2)

ghz.cx(6,7)
ghz.cx(7,8)

ghz.cx(2,1)
ghz.cx(2,0)

ghz.cx(8,9)
ghz.cx(8,10)



display(ghz.draw('mpl'))
Statevector(ghz).draw('latex')
```

```python
transpiled = transpiler.run(ghz)
print('final depth: ', transpiled.depth(depth))
transpiled.draw('mpl', idle_wires=False)
```
